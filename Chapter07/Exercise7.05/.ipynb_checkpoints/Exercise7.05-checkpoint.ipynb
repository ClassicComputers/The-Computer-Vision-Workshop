```python
import os 
from imutils import paths 
import cv2 
import numpy as np 
import matplotlib.pyplot as plt 
from skimage import feature 
from sklearn.svm import LinearSVC
```

```python
numPoints = 24
radius = 8
```

```python
data = []
labels = []
```

```python
from sklearn.svm import LinearSVC
from imutils import paths
import argparse
import cv2
import os
```

```python
data_path = 'lbp_data/training'
```

```python
os.listdir(data_path)
```

```python
eps = 1e-7
```

```python
paths.list_images(data_path)
```

```python
for imagePath in paths.list_images(data_path):
    # load the image, convert it to grayscale, and describe it
    image = cv2.imread(imagePath)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    lbp = feature.local_binary_pattern(gray, numPoints, radius, method="uniform")
    (hist, _) = np.histogram(lbp.ravel(), bins=np.arange(0, numPoints + 3), range=(0, numPoints + 2))
    hist = hist.astype("float")
    hist /= (hist.sum() + eps)
 
    # extract the label from the image path, then update the
    # label and data lists
    labels.append(imagePath.split(os.path.sep)[-2])
    data.append(hist)
 
    # train a Linear SVM on the data
model = LinearSVC(C=40, random_state=42)
model.fit(data, labels)
```

```python
data_path_test = 'lbp_data/test'
```

```python
test_data = []
pred = []
for imagePath in paths.list_images(data_path_test):
    # load the image, convert it to grayscale, describe it,
    # and classify it
    image = cv2.imread(imagePath)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    lbp = feature.local_binary_pattern(gray, numPoints, radius, method="uniform")
    (hist, _) = np.histogram(lbp.ravel(), bins=np.arange(0, numPoints + 3), range=(0, numPoints + 2))
    hist = hist.astype("float")
    hist /= (hist.sum() + eps)
    test_data.append(hist)
    print(hist)
    prediction = model.predict(hist.reshape(1, -1))
    pred.append(prediction)

    # display the image and the prediction
    cv2.putText(image, prediction[0], (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255))
    cv2.imshow("predicted image", image)
    cv2.waitKey(10000)
cv2.destroyAllWindows()
```
