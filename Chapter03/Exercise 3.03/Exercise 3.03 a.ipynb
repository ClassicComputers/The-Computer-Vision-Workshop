```python
import cv2
import matplotlib.pyplot as plt
img = cv2.imread('nature-branch-blossom-light-plant-photography-1361811-pxhere.com.jpg')
#img= cv2.imread('pexels-photo-1379636.jpeg')
origrgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB);
imgplot = plt.imshow(origrgb)
plt.title('Original')
plt.show()

imgLAB = cv2.cvtColor(img,cv2.COLOR_BGR2LAB);
imgLAB[:,:,0]= cv2.equalizeHist(imgLAB[:,:,0])
RGB = cv2.cvtColor(imgLAB,cv2.COLOR_LAB2RGB);
    
cv2.imwrite('LAB HE.png',cv2.cvtColor(imgLAB,cv2.COLOR_LAB2BGR))

imgplot = plt.imshow(RGB)
plt.title('HE in LAB')
plt.show()
```
